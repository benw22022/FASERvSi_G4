# How HepMC files are structured
Each HepMC file contains a list of events and each event has a list of vertices

Each vertex has two lists of attached particles - those going in and those going out

Each particle has the following properties'
- pid: The PDG particle Code
- momentum: The particle 4-momentum
- status: A number between 1 - 4. 1 = stable final state particle, 2 = intermediate decayed particle, 3 = undefined, 4 = initial state particle

```python
with pyhepmc.open("../GenieOutput/FASERvSi/FASERvSi.gfaser.part.0.hepmc") as f:
    for i, event in enumerate(f):
        print(f"\n{'='*100}")
        print(f"Event: {i}")
        for vertex in event.vertices:
            for particle in vertex.particles_in:
                momentum = particle.momentum
                print(f"In Particle: {get_name_from_pdgc(particle.pid):<30} Momentum: ({momentum.px:<8.3f}, {momentum.py:<8.3f}, {momentum.pz:<8.3f})    Staus: {particle.status}")
            print(f"{'-'*100}")    
            for particle in vertex.particles_out:
                momentum = particle.momentum
                print(f"Out Particle: {get_name_from_pdgc(particle.pid):<30} Momentum: ({momentum.px:<8.3f}, {momentum.py:<8.3f}, {momentum.pz:<8.3f})    Staus: {particle.status}")
        if i > 10:
            break
```
